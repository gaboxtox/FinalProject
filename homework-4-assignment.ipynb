## Import both data sets

### \(worth 10%, or 5% per data set\)

Write Python code below that loads the two data sets into this notebook and displays both\.

Note: This is one of those times where you should NOT use `print()` or `df.head()`\.

If your data sets need any cleaning or other manipulation done to prepare them for the merge in the next section, you may do that here as well\.  Use as many code cells as your work needs\.

```python
# Place your code here.  Use more than one cell if needed.
import pandas as pd

bls = pd.read_csv("bls_data.csv")
crime = pd.read_csv("CIUS_Crime_in_the_United_States_byState_2024.csv")

bls = bls.rename(columns={
    "FIPS Code": "fips_code",
    "State": "state",
    "Year": "year",
    "Civilian non-institutional population": "civilian_noninstitutional_population",
    "Total": "civilian_labor_force_total",
    "Percent of population": "civilian_labor_force_pct",
    "Total_Employment": "total_employment",
    "Percent of population.1": "employment_pct",
    "Total_Unemployment": "total_unemployment",
    "Rate": "unemployment_rate"
})

bls["state"] = bls["state"].str.strip()
crime["state"] = crime["state"].str.strip()

bls["year"] = pd.to_numeric(bls["year"], errors="coerce")
crime["year"] = pd.to_numeric(crime["year"], errors="coerce")

bls
crime
```

## Merge the two data sets

### \(worth 20%\)

```python
# Place your code here.  Use more than one cell if needed.
merged_df = pd.merge(
    bls,
    crime,
    on=["state", "year"],
    how="inner"
)

merged_df
```
